```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import pathlib
from datetime import datetime

from scipy.optimize import curve_fit
import natural_cubic_spline_stack_overflow_kopi as ncs # taken from stack overflow
```

```python
population = pd.read_excel("pop_regression_values_updated.xlsx")
PPD = pd.read_excel("PPD_predicted_values_first.xlsx")
```

```python
households = population["Predicted Population"] / PPD["Predicted_PPD"]

print(households)

households_df = pd.DataFrame(households, columns=["Predicted_Households"])
households_df["Year"] = PPD["Year"]
households_df = households_df[["Year", "Predicted_Households"]]
print(households_df)
```

```python
#plot households_df from 1600-2100
plt.figure(figsize=(10, 6))
plt.plot(households_df["Year"], households_df["Predicted_Households"], label="Households", color="orange")
plt.xlim(1600, 2100)
plt.xlabel("Year")
plt.ylabel("Count")
plt.title("Predicted No. of Households")
plt.legend()
plt.show()
```
